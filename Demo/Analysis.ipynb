```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Loading the Processed Data

```python
file_path = 'processed_data.csv'
df = pd.read_csv(file_path)
```

Total Number of Crimes

```python
# Total number of crimes (assuming each row represents a crime incident)
total_crimes = df.shape[0]
print(f"Total number of crimes: {total_crimes}")
```

Crimes in Each State

```python
# Group by state and count the number of crimes
crimes_by_state = df.groupby('State').size().reset_index(name='Number of Crimes')

# Display the top rows of crimes by state
print(crimes_by_state.head())

# Plotting the number of crimes by state using Seaborn
plt.figure(figsize=(12, 6))
sns.barplot(x='State', y='Number of Crimes', data=crimes_by_state, palette='coolwarm')
plt.xticks(rotation=90)
plt.title('Number of Crimes by State')
plt.show()
```

Crimes in Each City

```python
# Group by city and count the number of crimes
crimes_by_city = df.groupby('City').size().reset_index(name='Number of Crimes')

# Display the top rows of crimes by city
print(crimes_by_city.head())

# Plotting the number of crimes by city using Seaborn
plt.figure(figsize=(12, 6))
sns.barplot(x='City', y='Number of Crimes', data=crimes_by_city.sort_values(by='Number of Crimes', ascending=False).head(10), palette='viridis')
plt.xticks(rotation=90)
plt.title('Top 10 Cities with Highest Number of Crimes')
plt.show()
```

Crimes Sorted by Frequency

```python
# Sort the crimes by number in descending order (e.g., for states)
crimes_by_state_sorted = crimes_by_state.sort_values(by='Number of Crimes', ascending=False)

# Display the sorted crimes by state
print(crimes_by_state_sorted.head())

# Visualizing the sorted number of crimes by state
plt.figure(figsize=(12, 6))
sns.barplot(x='State', y='Number of Crimes', data=crimes_by_state_sorted, palette='Spectral')
plt.xticks(rotation=90)
plt.title('States Sorted by Number of Crimes')
plt.show()
```

Crimes by Category

```python
# Group by crime type (if available)
crimes_by_type = df.groupby('Crime Type').size().reset_index(name='Number of Crimes')

# Display the top crime types
print(crimes_by_type.head())

# Plotting the number of crimes by type
plt.figure(figsize=(12, 6))
sns.barplot(x='Crime Type', y='Number of Crimes', data=crimes_by_type, palette='Set2')
plt.xticks(rotation=90)
plt.title('Number of Crimes by Type')
plt.show()
```
